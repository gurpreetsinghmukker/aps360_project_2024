```python

def visualize_mel_spectrogram(mel_spectrogram, width=10, show_axes=True):
    fig = plt.figure(figsize=(width, 4))
    ax = fig.add_subplot(111)
    ax.imshow(mel_spectrogram[0], aspect='auto', origin='lower', cmap='viridis')

    if not show_axes:
        ax.set_axis_off()

    plt.show()
    plt.close()
```
